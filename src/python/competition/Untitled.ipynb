```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

data1 = np.loadtxt('data2.csv', delimiter=',')
data2 = np.loadtxt('data3.csv', delimiter=',')
input1 = data1[:,0]
output11 = data1[:,1]
output21= data2[:,1]
output12 = data1[:,2]
output22= data2[:,2]

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

x = np.linspace(-6,6,1000)

ax1.plot(input1, output11, color='pink', linestyle='solid', label='A(average)')
ax1.plot(input1, output21, color='lightblue', linestyle='solid', label='C(average)')
ax1.plot(input1, output12, color='red', linestyle='solid',label='A(max)')
ax1.plot(input1, output22, color='blue', linestyle='solid',label='C(max)')

ax1.legend()
ax1.set_title('ga mario')
ax1.set_xlabel('generation')
ax1.set_ylabel('reward')
ax1.grid(True)
```
